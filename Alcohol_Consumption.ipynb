### 1. Which continent drinks more beer on average?

```python
df.groupby('continent')['beer_servings'].mean().sort_values(ascending = False).reset_index()
```

### 2. For each continent print the statistics for wine consumption.

```python
df.groupby('continent')['wine_servings'].describe()
```

### 3. Print the mean alcohol consumption per continent for every column.

```python
df.groupby('continent')[['beer_servings', 'spirit_servings', 'wine_servings','total_litres_of_pure_alcohol']].mean().reset_index()
```

### 4. Print the median alcohol consumption per continent for every column.

```python
df.groupby('continent')[['beer_servings', 'spirit_servings', 'wine_servings','total_litres_of_pure_alcohol']].median().reset_index()
```

### Print the mean, min and max values for spirit consumption per Continent.

```python
df.groupby('continent')['spirit_servings'].agg(['mean', 'min', 'max'])
```
