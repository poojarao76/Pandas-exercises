```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data')
df
```

```python
col = df.iloc[:,[0,3,6,8,10,12,13]]
col
```

```python
df = df.drop(columns = col)
df
```

### 2. Assign the columns as below: - alcohol - malic_acid - alcalinity_of_ash - magnesium - flavanoids - proanthocyanins - hue.

```python
df.columns = ['alcohol', 'malic_acid', 'alcalinity_of_ash', 'magnesium', 'flavanoids', 'proanthocyanins', 'hue']
df.head()
```

### 3. Set the values of the first 3 rows in the alcohol column as NaN.

```python
df.loc[:2, 'alcohol'] = np.nan
df.head(10)
```

### 4. Now set the value of the rows 3 and 4 of the magnesium column as NaN

```python
df.loc[2:3, 'magnesium'] = np.nan
df.head(10)
```

### 5. Fill in the null values (NaN) with the number 10 in the alcohol column and 100 in magnesium column.

```python
df['alcohol'] = df['alcohol'].replace(np.nan, 10)
df.head()
```
